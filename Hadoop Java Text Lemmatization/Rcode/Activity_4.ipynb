```r
library(dplyr)
library(ggplot2)
library(reshape2)
```

```r
yax2 = c(202,279,323,399,444,677)
xax2 = c(10,15,20,25,30,40)
```

```r
df2 <- data.frame(xax2,yax2)
```

## Plot of Number of latin files used vs Running Time for Bigram

Conclusion - Here we see that Running time increases almost linearly with increase in number of latin files read

```r
plot(df2,type = 'o',col = "red",xlab = "",ylab = "")
title(main="Number of latin files vs Running Time in seconds - Bigram", col.main="green", font.main=4)
title(xlab="Number of latin files", col.lab=rgb(0,0.5,0))
title(ylab="Running Time in seconds", col.lab=rgb(0,0.5,0))
```
